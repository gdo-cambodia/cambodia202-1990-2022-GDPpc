```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```
url = 'https://github.com/gdo-cambodia/cambodia202-1990-2022-GDPpc/raw/refs/heads/main/cambodia202-1990-2022-GDPpc.dta'
response = requests.get(url)

# Save the content to a local file
with open('cambodia_gdp.dta', 'wb') as f:
    f.write(response.content)

# Read the .dta file into a pandas DataFrame
df = pd.read_stata('cambodia_gdp.dta')
```

```
df.head()
```

```
# Plot a box plot of 'y' by year
plt.figure(figsize=(15, 8))
sns.boxplot(x='year', y='y', data=df)
plt.title('Distribution of y by Year')
plt.xlabel('Year')
plt.ylabel('y')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(15, 8))
sns.boxplot(x='year', y='ln_y', data=df)
plt.title('Distribution of y by Year')
plt.xlabel('Year')
plt.ylabel('ln_y')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(15, 8))
sns.boxplot(x='year', y='trend400_ln_y', data=df)
plt.title('Distribution of y by Year')
plt.xlabel('Year')
plt.ylabel('trend400_ln_y')
plt.xticks(rotation=90)
plt.show()
```
